```python
# Dependencies
import numpy as np
import pandas as pd
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
# Database Setup
engine = create_engine("sqlite:///hawaii.sqlite")

# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)

# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

## D1: Determine the Summary Statistics for June

```python
# 1. Import the sqlalchemy extract function.
from sqlalchemy import extract

# 2. Write a query that filters the Measurement table to retrieve the temperatures for the month of June. 
june_temp = session.query(Measurement.date, Measurement.tobs).filter(extract('month', Measurement.date) == 6).all()
```

```python
#  3. Convert the June temperatures to a list.
print(june_temp)
```

```python
# 4. Create a DataFrame from the list of temperatures for the month of June. 
june_temp_df = pd.DataFrame(june_temp, columns=['date','June temps'])
june_temp_df.set_index(june_temp_df['date'], inplace=True)
june_temp_df = june_temp_df.sort_index()
print(june_temp_df.to_string(index=False))
```

```python
# 5. Calculate and print out the summary statistics for the June temperature DataFrame.
june_temp_df.describe()
```

## D2: Determine the Summary Statistics for December

```python
# 6. Write a query that filters the Measurement table to retrieve the temperatures for the month of December.
dec_temp = session.query(Measurement.date, Measurement.tobs).filter(extract('month', Measurement.date) == 12).all()
```

```python
# 7. Convert the December temperatures to a list.
print(dec_temp)
```

```python
# 8. Create a DataFrame from the list of temperatures for the month of December. 
dec_temp_df = pd.DataFrame(dec_temp, columns=['date','December temps'])
dec_temp_df.set_index(dec_temp_df['date'], inplace=True)
dec_temp_df = dec_temp_df.sort_index()
print(dec_temp_df.to_string(index=False))
```

```python
# 9. Calculate and print out the summary statistics for the Decemeber temperature DataFrame.
dec_temp_df.describe()
```

## D3: Statistical Analysis of Weather in June vs. December

```python
# Import dependencies for addional queries
%matplotlib inline
from matplotlib import style
style.use('fivethirtyeight')
import matplotlib.pyplot as plt
import datetime as dt
```

```python
# Plot June temps data
june_temp_df.plot.hist(bins=15)
plt.tight_layout()
plt.title('June Temperatures')
plt.xlim([65, 85])
plt.ylim([0,400])
```

```python
# Plot December temps data
dec_temp_df.plot.hist(bins=15)
plt.tight_layout()
plt.title('December Temperatures')
plt.xlim([55,85])
plt.ylim([0,400])
```

```python
# Additional Query 1: June Precipitation
june_temp_prcp = session.query(Measurement.date, Measurement.prcp).filter(extract('month', Measurement.date) == 6).all()
june_temp_prcp_df = pd.DataFrame(june_temp_prcp, columns=['date','December Precipitation'])
# print(june_temp_prcp_df)
june_temp_prcp_df.set_index(june_temp_prcp_df['date'], inplace=True)
june_temp_prcp_df = june_temp_prcp_df.sort_index()
print(june_temp_prcp_df.to_string(index=False))
```

```python
# Additional Query 2: December Precipitation 
dec_temp_prcp = session.query(Measurement.date, Measurement.prcp).filter(extract('month', Measurement.date) == 12).all()
dec_temp_prcp_df = pd.DataFrame(dec_temp_prcp, columns=['date','December Precipitation'])
# print(dec_temp_prcp_df)
dec_temp_prcp_df.set_index(dec_temp_prcp_df['date'], inplace=True)
dec_temp_prcp_df = dec_temp_prcp_df.sort_index()
print(dec_temp_prcp_df.to_string(index=False))
dec_temp_prcp_df.describe()
```
